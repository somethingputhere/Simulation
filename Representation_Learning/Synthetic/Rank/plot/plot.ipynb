```python
def prepare_plot_data(CR_Trials, iterations, trials):

    x_value = np.array([i for i in range(iterations+1)])
    y_value = np.zeros((trials, iterations+1))
    y_value = np.array([CR_Trials[T][1][:iterations+1] for T in range(trials)])

    return x_value, np.mean(y_value, axis=0), np.std(y_value, axis=0) / np.sqrt(trials)
```

```python
r = [2, 4, 6]
trials = 100
iterations = 200
every_point = 25
```

```python
CR_trials_2 = np.load('../r = 2/CR_Trials_ICML.npy', allow_pickle=True)
CR_trials_4 = np.load('../r = 4/CR_Trials_ICML.npy', allow_pickle=True)
CR_trials_6 = np.load('../r = 6/CR_Trials_ICML.npy', allow_pickle=True)
```

```python
x_2, CR_2, std_2 = prepare_plot_data(CR_trials_2, iterations, trials)
x_4, CR_4, std_4 = prepare_plot_data(CR_trials_4, iterations, trials)
x_6, CR_6, std_6 = prepare_plot_data(CR_trials_6, iterations, trials)
```

```python
plt.figure(figsize=(12, 8), dpi = 300)

colors = (['black', 'blue', 'darkgreen', 'purple', 'darkred', 'grey'])
markers = ['*', 's', 'o', 'X', '^', 'P']

plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rc('legend', fontsize = 25)

plt.plot(x_2, CR_2, color = colors[0], linewidth = 3, label = 'r = 2')
plt.scatter(x_2[::every_point], CR_2[::every_point], marker = markers[0], color = colors[0], s = 300, label = 'r = 2')
plt.plot(x_4, CR_4, color = colors[1], linewidth = 3, label = 'r = 4')
plt.scatter(x_4[::every_point], CR_4[::every_point], marker = markers[1], color = colors[1], s = 300, label = 'r = 4')
plt.plot(x_6, CR_6, color = colors[2], linewidth = 3, label = 'r = 6')
plt.scatter(x_6[::every_point], CR_6[::every_point], marker = markers[2], color = colors[2], s = 300, label = 'r = 6')

legend_elements = [mlines.Line2D([0], [0], color = colors[0], lw = 5, label = 'r = 2', marker = markers[0], markersize = 15),
                   mlines.Line2D([0], [0], color = colors[1], lw = 5, label = 'r = 4', marker = markers[1], markersize = 15),
                   mlines.Line2D([0], [0], color = colors[2], lw = 5, label = 'r = 6', marker = markers[2], markersize = 15)]

plt.grid(True)
plt.xlabel('Round, N', fontsize = 25)
plt.ylabel(r'Cumulative Regret, $R_{T, n}$', fontsize = 25)
plt.legend(handles=legend_elements)
plt.title('Synthetic data: d = 100, T = 100', fontsize = 25)
# plt.savefig('synthetic_cumulative_regret_rank.pdf', dpi = 600, bbox_inches = 'tight')
plt.show()
plt.close()
```
